Dicionário de Dados Completo (Usei o IA para entender o dado real de cada coluna)

1. Identificadores Básicos (Dados Cadastrais)

***ID: Identificador único de cada registro/linha na tabela.
***Customer_ID: Identificador único do cliente (um cliente pode se repetir em meses diferentes).
***Month: Mês de referência daquela coleta de dados (Ex: January, February).
***Name: Nome completo do cliente.
***Age: Idade do cliente.
***SSN: Social Security Number (equivalente ao CPF nos Estados Unidos).
***Occupation: Profissão ou ocupação do cliente (Ex: Scientist, Engineer).


2. Indicadores Financeiros (Renda e Saldos)

***Annual_Income: Renda anual declarada pelo cliente.
***Monthly_Inhand_Salary: Salário mensal líquido (o valor que cai "na mão" após descontos).
***Monthly_Balance: Saldo mensal restante na conta bancária após todos os gastos e investimentos.
***Amount_invested_monthly: Valor investido mensalmente pelo cliente.


3. Produtos e Histórico Bancário

***Num_Bank_Accounts: Quantidade de contas bancárias ativas que o cliente possui.
***Num_Credit_Card: Quantidade de cartões de crédito que o cliente possui.
***Interest_Rate: Taxa de juros aplicada aos cartões de crédito do cliente.
***Num_of_Loan: Quantidade de empréstimos ativos que o cliente pegou no banco.
***Type_of_Loan: Tipos de empréstimos contratados (Ex: Home Loan, Car Loan, Credit Builder Loan).


4. Comportamento de Crédito e Débitos

***Delay_from_due_date: Média de dias de atraso no pagamento das faturas em relação à data de vencimento.
***Num_of_Delayed_Payment: Número médio de pagamentos atrasados pelo cliente.
***Changed_Credit_Limit: Percentual de alteração/mudança no limite do cartão de crédito do cliente.
***Num_Credit_Inquiries: Número de consultas ao crédito do cliente (quantas vezes bureaus de crédito foram acionados).
***Credit_Mix: Classificação do mix de créditos do cliente (a variedade de produtos financeiros utilizados. Ex: Good, Standard, Bad).
***Outstanding_Debt: Dívida restante total acumulada a ser paga (em dólares).
***Credit_Utilization_Ratio: Taxa de utilização de crédito (o percentual de limite do cartão que o cliente de fato gasta).
***Credit_History_Age: Tempo/Idade do histórico de crédito do cliente (geralmente escrito em formato de texto como "22 Years and 1 Months").
***Payment_of_Min_Amount: Indica se o cliente pagou apenas o valor mínimo da fatura do cartão (Yes ou No).
***Total_EMI_per_month: Valor total pago mensalmente em prestações fixas/parcelas de empréstimos (Equated Monthly Installment).
***Payment_Behaviour: Comportamento de pagamento do cliente (Ex: High_spent_Small_value_payments - Gasta muito em pagamentos de valores pequenos).

5. Variável Alvo

***Credit_Score: A classificação final do score de crédito do cliente dada pelo banco. Ela é dividida em 3 classes: Poor (Ruim), Standard (Médio) ou Good (Bom).

```python
# Valores Ausentes
print(df.isnull().sum())
```

```python
# Substitui os nulos de 'Name' por 'Desconhecido'
df['Name'] = df['Name'].fillna('Desconhecido')
```

```python
# Substitui os nulos de 'Type_of_Loan' por 'Não Informado'
df['Type_of_Loan'] = df['Type_of_Loan'].fillna('Não Informado')
```

```python
# Salário Mensal Líquido: Preenchi com a Mediana
mediana_salario = df['Monthly_Inhand_Salary'].median()
df['Monthly_Inhand_Salary'] = df['Monthly_Inhand_Salary'].fillna(mediana_salario)
```

```python
# Tive a ajuda da IA para conseguir concluir essa parte, pois a substituição dos valores estava dando erro, por conta dos caracteres das colunas
# Colunas financeiras mais importantes que estão dando erro, por conter caracteres diferentes
colunas_financeiras = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Amount_invested_monthly', 'Monthly_Balance', 'Outstanding_Debt', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries']

for col in colunas_financeiras:
    # Transforma tudo em texto e limpa os caracteres especiais
    df[col] = df[col].astype(str).str.replace(r'[^0-9.]', '', regex=True)
    # Converte de texto para número (float/decimal)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Para as colunas numéricas, realizei o calculo da mediana para preencher os valores vazios/nulos
for col in colunas_financeiras:
    mediana_da_coluna = df[col].median()
    df[col] = df[col].fillna(mediana_da_coluna)

# Para as colunas de texto
df['Credit_History_Age'] = df['Credit_History_Age'].fillna('Não Informado')
```

```python
# Atualização se ainda há valores a serem tratados
print(df.isnull().sum())
```

```python
df.head()
```

```python
# Verificando Duplicidade
linhas_duplicadas = df.duplicated().sum()
print (linhas_duplicadas)
```

```python
# Colunas que envolvem dinheiro ou valores decimais longos
colunas_para_arredondar = [
    'Annual_Income',
    'Monthly_Inhand_Salary',
    'Outstanding_Debt',
    'Amount_invested_monthly',
    'Monthly_Balance',
    'Total_EMI_per_month',
    'Credit_Utilization_Ratio'
]

# Loop para aplicar o arredondamento em cada coluna da lista
for col in colunas_para_arredondar:
    if col in df.columns:
        df[col] = df[col].round(2)
```

```python
# Utilizando apenas a informação de ano da Coluna Credit_History_Age - Corta o texto pelo primeiro espaço em branco e pega apenas a primeira parte [0]
df['Credit_History_Age'] = df['Credit_History_Age'].astype(str).str.split(' ').str[0]

# Se o valor for "nan", vazio ou não for um dígito numérico, substui por "Não Informado", para não gerar uma analise equivocada dos dados
df['Credit_History_Age'] = df['Credit_History_Age'].apply(lambda x: x if str(x).isdigit() else 'Não Informado')
```

```python
df.head(150)
```

```python
# Simplificando dados da Coluna Payment_Behaviour
dados_coluna = {
    'Low_spent_Small_value_payments': 'Gasto Baixo - Parcelas Pequenas',
    'Low_spent_Medium_value_payments': 'Gasto Baixo - Parcelas Médias',
    'Low_spent_Large_value_payments': 'Gasto Baixo - Parcelas Altas',
    'High_spent_Small_value_payments': 'Gasto Alto - Parcelas Pequenas',
    'High_spent_Medium_value_payments': 'Gasto Alto - Parcelas Médias',
    'High_spent_Large_value_payments': 'Gasto Alto - Parcelas Altas'
}


df['Payment_Behaviour'] = df['Payment_Behaviour'].map(dados_coluna)

# Como a coluna possui ruídos de texto (como "!@9#%"), preencho os dados Nulos com "Não Informado"
df['Payment_Behaviour'] = df['Payment_Behaviour'].fillna('Não Informado')

print(df['Payment_Behaviour'].value_counts())
```

### TRATAMENTO DE DADOS DAS COLUNAS

```python
df.describe()
```

```python
# 2. Tratamento das Contas Bancárias com Mediana
df.loc[(df['Num_Bank_Accounts'] < 0) | (df['Num_Bank_Accounts'] > 10), 'Num_Bank_Accounts'] = 6

# 3. Tratamento de Cartões com Mediana
df.loc[df['Num_Credit_Card'] > 12, 'Num_Credit_Card'] = 5

# 4. Tratamento de dias negativos
df.loc[df['Delay_from_due_date'] < 0, 'Delay_from_due_date'] = 0

# 5. Tratamento contagens de atrasos e consultas com Mediana
df.loc[df['Num_of_Delayed_Payment'] > 30, 'Num_of_Delayed_Payment'] = 14
df.loc[df['Num_Credit_Inquiries'] > 30, 'Num_Credit_Inquiries'] = 6

# 6. Tratamento do Saldo Mensal - valor fora  da realidade = mediana
mediana_saldo = df['Monthly_Balance'].median()
df.loc[df['Monthly_Balance'] > 100000, 'Monthly_Balance'] = mediana_saldo
```

```python
df.describe()
```

```python
# Mediana Renda
mediana_renda = df['Annual_Income'].median()

# Se a renda anual for maior que 200.000, substitui pela mediana
df.loc[df['Annual_Income'] > 200000, 'Annual_Income'] = mediana_renda

# Arredondamento para duas casas decimais
df['Annual_Income'] = df['Annual_Income'].round(2)
```

### INSIGTHS VISUAIS (GRÁFICOS)

```python
import matplotlib.pyplot as plt
```

```python
# Divida Media por Cartão  vs Score de Credito
df['Divida_Por_Cartao'] = df['Outstanding_Debt'] / df['Num_Credit_Card'].replace(0, 1)

print("--- INSIGHT 1: Alavancagem de Risco por Categoria de Score ---")
analise_cartao = df.groupby('Credit_Score')['Divida_Por_Cartao'].mean().round(2)
print(analise_cartao, "\n")

plt.figure(figsize=(7, 4))
analise_cartao.plot(kind='bar', color=['#e74c3c', '#2ecc71', '#f1c40f'])
plt.title('Divida Media por Cartão  vs Score de Credito')
plt.xlabel('Classificação do Score')
plt.ylabel('Valor Divida por Cartão')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Distribuição de Renda Média por Comportamento de Pagamento
analise_renda_comportamento = df.groupby('Payment_Behaviour')['Annual_Income'].mean().sort_values().round(2)
print(analise_renda_comportamento, "\n")

plt.figure(figsize=(8, 4))
analise_renda_comportamento.plot(kind='barh', color='#34495e')
plt.title('Renda Anual Media por Perfil de Comportamento')
plt.xlabel('Renda Anual Média ($)')
plt.ylabel('Perfil de Pagamento')
plt.tight_layout()
plt.show()
```

```python
# Dias de Atraso Médio por Faixa de Score
analise_atraso = df.groupby('Credit_Score')['Delay_from_due_date'].mean().round(1)
print(analise_atraso, "\n")

plt.figure(figsize=(7, 4))
analise_atraso.plot(kind='line', marker='o', linewidth=2, color='#9b59b6')
plt.title('Dias Médios de Atraso por Categoria de Score')
plt.xlabel('Classificação do Score')
plt.ylabel('Média de Dias de Atraso do Vencimento')
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# Renda Média X  Profissão

profissao_salario = df.groupby('Occupation')['Monthly_Inhand_Salary'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
profissao_salario.plot(kind='barh', color='#2ca02c')

plt.title('Salário Mensal Medio por Profissão', fontsize=14, fontweight='bold')
plt.xlabel('Salário Mensal Medio', fontsize=12)
plt.ylabel('Profissão', fontsize=12)
plt.tight_layout()
plt.show()
```

*Observação: Para que o projeto fluisse, as idades inconsistentes que vieram na base original foram preenchidas com a idade mediana, que é 34 anos.*


```python
# Criando faixa etária para organizar os dados
def criar_faixas(idade):
    if idade < 25:
        return '18 a 24 anos'
    elif idade < 35:
        return '25 a 34 anos'
    elif idade < 45:
        return '35 to 44 anos'
    elif idade < 55:
        return '45 a 54 anos'
    else:
        return '55 anos ou mais'

# Aplica a função criando uma nova coluna chamada 'Faixa_Etaria'
df['Faixa_Etaria'] = df['Age'].apply(criar_faixas)
```

```python
tabela_cruzada = pd.crosstab(df['Faixa_Etaria'], df['Credit_Score'], normalize='index') * 100

# Gerando o Gráfico de Colunas Empilhadas
plt.figure(figsize=(9, 5))
tabela_cruzada.plot(kind='bar', stacked=True, color=['#2ecc71', '#e74c3c', '#f1c40f'], ax=plt.gca())

plt.title('Proporção de Score de Crédito por Faixa Etária', fontsize=14, fontweight='bold')
plt.xlabel('Faixas de Idade Criadas', fontsize=12)
plt.ylabel('Percentual do Total (%)', fontsize=12)
plt.legend(title='Score de Crédito', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Faixa Etaria por Score

tabela_prova = pd.crosstab(df['Faixa_Etaria'], df['Credit_Score'], normalize='index') * 100

plt.figure(figsize=(8, 4))
plt.plot(tabela_prova.index, tabela_prova['Poor'], marker='o', color='#e74c3c', linewidth=3)

plt.title('Prova Real: Indice de Risco (Score Poor) por Faixa Etária')
plt.xlabel('Faixas de Idade')
plt.ylabel('Percentual de Clientes Ruins (%)')
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```
